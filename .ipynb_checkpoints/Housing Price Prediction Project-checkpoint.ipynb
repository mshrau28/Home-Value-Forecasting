Geospatial Analysis

```python
city_center = (40.7128, -74.0060)

# Calculate distance from city center
def calculate_distance(row):
    return geodesic(city_center, (row['latitude'], row['longitude'])).kilometers  

df['distance_from_center'] = df.apply(calculate_distance, axis=1)

# Plotting listings on a map using Basemap 
plt.figure(figsize=(14, 10))

# Create a map centered around the data with a larger zoom level
m = Basemap(projection='merc', llcrnrlat=40.30, urcrnrlat=41.00,  # Adjusted latitude bounds 
            llcrnrlon=-74.26, urcrnrlon=-73.60, resolution='i')  # Adjusted longitude bounds 

m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.drawmapboundary()
m.drawrivers()
m.fillcontinents(color='lightgray', lake_color='aqua')

# Plot each listing
for idx, row in df.iterrows():
    x, y = m(row['longitude'], row['latitude'])
    color = 'blue' if row['log_price'] < 4.5 else 'red'
    m.plot(x, y, 'o', markersize=8, color=color, alpha=0.75)

plt.title('Listing Prices on Map')
plt.show()

# Plot price relation by neighbourhood
plt.figure(figsize=(12, 6))
plt.scatter(df['neighbourhood'], df['log_price'])
plt.title('Price Relation by Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('log_price')
plt.show()

# Plot price against distance from city center
plt.figure(figsize=(12, 6))
sns.scatterplot(x='distance_from_center', y='log_price', hue='log_price', data=df)
plt.title('Price Distribution by Distance from City Center')
plt.xlabel('Distance from City Center (km)')
plt.ylabel('log_price')
plt.legend(title='log_price', loc='upper right')
plt.show()
```

Sentiment Analysis on Textual Data

```python
# Initialize WordNet Lemmatizer and retrieve English stopwords
lemma = WordNetLemmatizer()
stop_words = stopwords.words('english')

# Define a function for text preprocessing
def text_prep(x: str) -> list:
    # Convert text to lowercase
    corp = str(x).lower()
    # Remove non-alphabetic characters and strip leading/trailing whitespaces
    corp = re.sub('[^a-zA-Z]+', ' ', corp).strip()
    # Tokenize the text
    tokens = word_tokenize(corp)
    # Remove stopwords
    words = [t for t in tokens if t not in stop_words]
    # Lemmatize the words
    lemmatize = [lemma.lemmatize(w) for w in words]
    return lemmatize

# Preprocess the 'description' column and create a new column 'preprocess_txt'
preprocess_tag = [text_prep(i) for i in df['description']]
df["preprocess_txt"] = preprocess_tag

# Open the files with the correct encoding to retrieve positive and negative words
with open('C:/Users/shram/Downloads/negative-words.txt', 'r', encoding='ISO-8859-1') as file:  #encoding indicates latin
    neg_words = file.read().split()

with open('C:/Users/shram/Downloads/positive-words.txt', 'r', encoding='ISO-8859-1') as file:
    pos_words = file.read().split()

# Count the number of positive words in each preprocessed text
num_pos = df['preprocess_txt'].map(lambda x: len([i for i in x if i in pos_words]))
df['pos_count'] = num_pos

# Count the number of negative words in each preprocessed text
num_neg = df['preprocess_txt'].map(lambda x: len([i for i in x if i in neg_words]))
df['neg_count'] = num_neg

# Calculate the sentiment score based on the ratio of positive to negative words
df['sentiment'] = round(df['pos_count'] / (df['neg_count'] + 1), 2)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Define a threshold value
threshold = 1  

# Function to determine sentiment label
def classify_sentiment(score):
    if score > threshold:
        return 'positive'
    else :  
        return 'negative'
# Apply the function to the 'sentiment_score' column to assign sentiment labels
df['sentiment_label'] = df['sentiment'].apply(classify_sentiment)
```

Train-Test Split

```python

# Assuming df is your DataFrame with all the features and the target variable
# X contains the features, y contains the target variable
X=X_encoded
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Create and fit the Random Forest regression model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print("Mean Squared Error:", mse)
print("R-squared score:", r2)


```

Optimization Hyperparameter Tuning

```python
X = X_encoded  
y = y 

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Initialize GridSearchCV with RandomForestRegressor
grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                           param_grid=param_grid,
                           cv=5,  # 5-fold cross-validation
                           n_jobs=-1,  # Use all available cores
                           verbose=2)  # Increase verbosity for more detailed output

grid_search.fit(X_train, y_train)

# Print the best parameters 
print("Best Parameters found by GridSearchCV:")
print(grid_search.best_params_)

# Use the best model to make predictions on the test set
best_model = grid_search.best_estimator_
predictions = best_model.predict(X_test)

# Evaluate the best model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Mean Squared Error:", mse)
print("R-squared score:", r2)
```
